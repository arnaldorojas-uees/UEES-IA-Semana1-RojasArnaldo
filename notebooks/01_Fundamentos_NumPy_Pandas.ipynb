## 1. NumPy: Manipulación de Arrays

NumPy (Numerical Python) es la librería fundamental para la computación numérica en Python. Proporciona un objeto de array multidimensional de alto rendimiento y herramientas para trabajar con estos arrays.

```python
import numpy as np

# Crear un array unidimensional
arr1 = np.array([1, 2, 3, 4, 5])
print("Array unidimensional:", arr1)
print("Tipo de dato:", arr1.dtype)
print("Dimensiones:", arr1.shape)

# Crear un array bidimensional (matriz)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\nArray bidimensional:\n", arr2)
print("Tipo de dato:", arr2.dtype)
print("Dimensiones:", arr2.shape)

# Crear un array con un rango de valores
arr_range = np.arange(10)
print("\nArray con rango:", arr_range)

# Crear un array de ceros y unos y valores aleatorios
zeros_arr = np.zeros((2, 3))
ones_arr = np.ones((3, 2))
randn = np.random.randn(3, 3)
print("\nArray de ceros:\n", zeros_arr)
print("\nArray de unos:\n", ones_arr)
print("\nArray de Randoms normal:\n", randn)
```

### Explicación de las operaciones básicas de NumPy

- `np.array()`: Se usa para crear arrays a partir de listas o tuplas de Python.
- `dtype`: Atributo que muestra el tipo de dato de los elementos del array.
- `shape`: Atributo que devuelve una tupla indicando las dimensiones del array.
- `np.arange()`: Crea arrays con valores espaciados uniformemente dentro de un intervalo.
- `np.zeros()` y `np.ones()`: Crean arrays de un tamaño dado, rellenos con ceros o unos, respectivamente.

Indexación y slicing (código)

```python
arr = np.array([10, 20, 30, 40, 50, 60])
print("Array completo:", arr)

# Acceder a un elemento
print("Elemento en índice 0:", arr[0])
print("Elemento en índice 3:", arr[3])

# Slicing
print("Elementos del 1 al 4 (sin incluir el índice 4):", arr[1:4])
print("Últimos 3 elementos:", arr[-3:])
```

## 2. Fundamentos de Pandas

Pandas proporciona dos estructuras principales:

- **Series**: similares a una columna.
- **DataFrame**: tabla con filas y columnas (como Excel, pero en código).

Trabajaremos primero con datos creados manualmente y luego con un dataset de prueba.


```python
# Crear una Serie
s = pd.Series([10, 20, 30, 40], name="ventas")
print("Serie s:")
print(s)

# Crear un DataFrame a partir de un diccionario
datos = {
    "producto": ["A", "B", "C", "D"],
    "precio": [10.5, 20.0, 7.5, 15.0],
    "cantidad": [100, 50, 80, 120]
}

df = pd.DataFrame(datos)
print("\nDataFrame df:")
print(df)
```

**Inspección básica del DataFrame**

```python
print("Primeras filas:")
display(df.head())

print("\nInformación general:")
print(df.info())

print("\nEstadísticas descriptivas:")
display(df.describe())
```

**Selección y filtrado**

```python
# Seleccionar una columna
print("Columna 'precio':")
print(df['precio'])

# Seleccionar varias columnas
print("\nColumnas 'producto' y 'cantidad':")
print(df[['producto', 'cantidad']])

# Filtrado por condición
print("\nProductos con precio > 10:")
print(df[df['precio'] > 10])

# Agregar columna calculada
df['total'] = df['precio'] * df['cantidad']
print("\nDataFrame con columna 'total':")
print(df)
```

**Agrupación (groupby)**

```python
# Supongamos categorías
df['categoria'] = ['X', 'X', 'Y', 'Y']

print("DataFrame con categoría:")
print(df)

# Agrupar por categoría y calcular estadísticas
grouped = df.groupby('categoria')['total'].sum()
print("\nTotal vendido por categoría:")
print(grouped)
```

**USANDO UN DATASET DE PRUEBA (TITANIC)**

```python
# Dataset Titanic desde seaborn
titanic = sns.load_dataset('titanic')

print("Primeras filas del dataset Titanic:")
display(titanic.head())

print("\nInformación general:")
print(titanic.info())

print("\nEstadísticas de columnas numéricas:")
display(titanic.describe())

# Seleccionar columnas relevantes
titanic_sub = titanic[['survived', 'pclass', 'sex', 'age', 'fare']]
print("Subset de columnas seleccionadas:")
display(titanic_sub.head())

# Filtrar pasajeros con tarifa mayor a 50
caros = titanic_sub[titanic_sub['fare'] > 50]
print("\nPasajeros con fare > 50:")
display(caros.head())

# Tasa de supervivencia por sexo
surv_by_sex = titanic.groupby('sex')['survived'].mean()
print("Tasa de supervivencia por sexo:")
print(surv_by_sex)

# Tasa de supervivencia por clase y sexo
surv_by_class_sex = titanic.groupby(['pclass', 'sex'])['survived'].mean()
print("\nTasa de supervivencia por clase y sexo:")
print(surv_by_class_sex)

# Contar valores faltantes por columna
missing = titanic.isnull().sum()
print("Valores faltantes por columna:")
print(missing)

# Ejemplo: eliminar filas con age nulo
titanic_age_clean = titanic.dropna(subset=['age'])
print("\nFilas después de eliminar nulos en 'age':", titanic_age_clean.shape[0])

# Ejemplo: rellenar nulos de 'age' con la media
titanic_fill_age = titanic.copy()
titanic_fill_age['age'] = titanic_fill_age['age'].fillna(titanic_fill_age['age'].mean())
print("¿Quedan nulos en 'age'?:", titanic_fill_age['age'].isnull().sum())

```

**USANDO UN DATASET DE PRUEBA (IRIS)**

```python
# Dataset Iris desde seaborn
iris = sns.load_dataset('iris')

print("Primeras filas del dataset Iris:")
display(iris.head())

print("\nInformación general:")
print(iris.info())

print("\nEstadísticas de columnas numéricas:")
display(iris.describe())

iris_sub = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']]
print("Subset de columnas seleccionadas:")
display(iris_sub.head())

# Filtrar flores con pétalo largo (> 5 cm)
petalos_largos = iris_sub[iris_sub['petal_length'] > 5]
print("\nFlores con petal_length > 5:")
display(petalos_largos.head())

# Media de las medidas por especie
mean_by_species = iris.groupby('species')[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].mean()
print("Medias de las medidas por especie:")
display(mean_by_species)

# Número de observaciones por especie
count_by_species = iris['species'].value_counts()
print("\nNúmero de muestras por especie:")
print(count_by_species)

# Matriz de correlación entre variables numéricas
corr_matrix = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()
print("Matriz de correlación:")
display(corr_matrix)

# (Opcional) Mapa de calor rápido
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlación entre variables del dataset Iris")
plt.show()
```
